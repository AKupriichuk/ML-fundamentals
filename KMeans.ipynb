Метод кластеризації KMeans за замовчуванням використовує Евклідову відстань як метрику відстані. Тому тобі не потрібно окремо вказувати метрику в KMeans, оскільки вона вже використовує саме її.

Зробимо візуалізацію силуету

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, confusion_matrix
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
from yellowbrick.cluster import silhouette_visualizer
from yellowbrick.datasets import load_credit

# Функція для кластеризації та візуалізації
def perform_clustering(X, clusters_range=range(3, 6)):
    # Стандартизуємо дані
    scaler = StandardScaler()
    scaled_X = scaler.fit_transform(X)

    # Для збереження коефіцієнтів "силуету" та відстаней
    silhouette_scores = []
    intra_cluster_distances = []
    inter_cluster_distances = []

    # Перебір кількості кластерів
    for k in clusters_range:
        # Кластеризація
        model = KMeans(n_clusters=k)
        model.fit(scaled_X)

        # Отримуємо метки кластерів
        labels = model.labels_

        # Обчислюємо коефіцієнт "силуету"
        sil_score = silhouette_score(scaled_X, labels, metric='euclidean')
        silhouette_scores.append(sil_score)

        # Обчислюємо внутрішньокластерну відстань
        intra_distance = np.mean([np.mean(pairwise_distances(scaled_X[labels == i], metric='euclidean')) for i in range(k)])
        intra_cluster_distances.append(intra_distance)

        # Обчислюємо міжкластерну відстань
        inter_distance = []
        for i in range(k):
            for j in range(i + 1, k):
                dist = np.mean(pairwise_distances(scaled_X[labels == i], scaled_X[labels == j], metric='euclidean'))
                inter_distance.append(dist)
        inter_cluster_distances.append(np.mean(inter_distance) if inter_distance else 0)

        # Виводимо кількість кластерів та силуэтний коефіцієнт
        print(f'Кількість кластерів: {k}, Коефіцієнт силуету: {sil_score}')
        print("\nВнутрішньокластерні відстані:", intra_cluster_distances)
        print("Міжкластерні відстані:", inter_cluster_distances)
        # Тривимірна візуалізація результатів кластеризації
        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(scaled_X[:, 0], scaled_X[:, 1], scaled_X[:, 2], c=labels, cmap='viridis', marker='o', edgecolor='k')
        ax.set_title(f'Кластеризація для {k} кластерів (KMeans)')
        ax.set_xlabel('Талия')
        ax.set_ylabel('Стегна')
        ax.set_zlabel('Вага')
        plt.show()
        silhouette_visualizer(KMeans(k, random_state=42), X, colors='yellowbrick')
    # Повертаємо кількість кластерів з найвищим коефіцієнтом "силуету"
    best_k = clusters_range[np.argmax(silhouette_scores)]
    print(f'Найбільш точне розбиття: {best_k} кластери з коефіцієнтом "силуету" {max(silhouette_scores)}')

    return best_k

# Виклик функції для проведення кластерного аналізу
best_clusters = perform_clustering(data)
```
